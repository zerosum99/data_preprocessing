```python
import matplotlib.pyplot as plt
```

### 데이터 만들기

```python
import numpy as np
```

```python
x = np.linspace(0,10, 100)
```

```python
y = np.cos(x)
```

```python
z = np.sin(x)
```

### 여러 개의 그래프를 하나로 꾸미기

```python
plt.plot(x, x, x, x**2, x, x**3) 
plt.show()
```

### 데이터 색상 조절하기

```python
plt.plot(x, y, alpha = 0.4)
plt.show()
```

```python
plt.plot(x, y, c='k') 
plt.show()
```

### 마커 표시하기 

```python
fig, ax = plt.subplots() 
ax.scatter(x,y,marker=".")  

plt.show()
```

```python
fig, ax = plt.subplots() 
ax.plot(x,y,marker="o")
plt.show()
```

### 라인스타일 처리하기 

```python
 plt.plot(x,y,linewidth=4.0) 
```

```python
plt.plot(x,y,ls='solid')
```

```python
 plt.plot(x,y,ls='--')
```

```python
plt.plot(x,y,'--',x**2,y**2,'-.') 
```

### setp 함수를 이용해서 꾸미기

```python
fig, ax = plt.subplots()
lines = ax.plot(x,y) 
plt.setp(lines,color='r',linewidth=4.0)
plt.show()
```

### 텍스트만 추가하기

```python
fig, ax = plt.subplots()
lines = ax.plot(x,y) 
plt.setp(lines,color='r',linewidth=4.0)
ax.text(1,  0.9,    'Example Graph',   style='italic') 
plt.show()
```

### 어노테이션 추가하기

```python
fig, ax = plt.subplots()
lines = ax.plot(x,y) 
plt.setp(lines,color='r',linewidth=4.0)
ax.text(1,  0.9,    'Example Graph',   style='italic') 
ax.annotate("Sine",   
             xy=(8, 0), 
             xycoords='data', 
             xytext=(10.5, 0),  
             textcoords='data',  
             arrowprops=dict(arrowstyle="->", 
                             connectionstyle="arc3"),)
plt.show()
```

### 그래프에 제목붙이기

```python
fig, ax = plt.subplots()
lines = ax.plot(x,y) 
plt.setp(lines,color='r',linewidth=4.0)
plt.title(r'$sigma_i=15$', fontsize=20)
plt.show()
```

###    Limits & Autoscaling 

```python
fig, ax = plt.subplots()
lines = ax.plot(x,y) 
ax.margins(x=0.0,y=0.1)  
ax.axis('equal')
```

```python
fig, ax = plt.subplots()
lines = ax.plot(x,y) 
ax.margins(x=0.0,y=0.1)  
ax.axis('equal')
ax.set_xlim(0,10.5)
```

```python
fig, ax = plt.subplots()
lines = ax.plot(x,y) 
ax.margins(x=0.0,y=0.1)  
ax.axis('equal')
ax.set(xlim=[0,10.5],ylim=[-1.5,1.5]) 
```

### 제목과 축 레이블 처리

```python
fig, ax = plt.subplots()
lines = ax.plot(x,y) 
ax.margins(x=0.0,y=0.1)  
ax.axis('equal')
ax.set(xlim=[0,10.5],ylim=[-1.5,1.5]) 

ax.set(title='An Example Axes',ylabel='Y-Axis',xlabel='X-Axis')
```

### 범례 표시하기

```python
fig, ax = plt.subplots()
lines = ax.plot(x,y) 
ax.margins(x=0.0,y=0.1)  
ax.axis('equal')
ax.set(xlim=[0,10.5],ylim=[-1.5,1.5]) 

ax.set(title='An Example Axes',ylabel='Y-Axis',xlabel='X-Axis')
ax.legend(['A simple line'],loc='best')
```

### 그래프에 범레를 추가 

```python

plt.plot(x,y,'--', label='x1')
plt.plot(x**2,y**2,'-.',label='x2') 
    
plt.legend(loc='best')
    
```

### 축 표시하기 

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4,5]) 
ax.xaxis.set(ticks=range(1,5),ticklabels=[3,100,-12,"foo"]) 
ax.tick_params(axis='y',  direction='inout',  length=10)
```

## 세부 그래프 간격 조정하기 

```python
fig3, axes = plt.subplots(nrows=2,ncols=2) 
fig3.subplots_adjust(wspace=0.5,              #Adjust the spacing between subplots   
                    hspace=0.3,left=0.125,   right=0.9,   top=0.9,  bottom=0.1) 

```

```python
fig2, axes = plt.subplots(nrows=2,ncols=2) 
fig2.tight_layout()   #Fit subplot(s) in to the figure area   Axis Spines 
```

### 좌표 측 조정하기 

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4,5]) 
fig.subplots_adjust(wspace=0.5,              #Adjust the spacing between subplots   
                    hspace=0.3,left=0.125,   right=0.9,   top=0.9,  bottom=0.1) 

fig.tight_layout()                            #Fit subplot(s) in to the figure area   Axis Spines 
ax.spines['top'].set_visible(False)          #Make the top axis line for a plot invisible 
ax.spines['bottom'].set_position(('outward',10)) # Move the bottom axis line outward
```

## 1. 선그래프 그리기

```python
import numpy as np
```

```python
x = np.linspace(0,10, 100)
```

### 그래프 그리기

```python
plt.plot(x)
```

### 좌표 축을 없애기 

```python
plt.plot(x)
plt.axis('off')
```

```python
y = np.cos(x)
```

```python
plt.plot(y)
```

```python
z = np.sin(x)
```

```python
plt.plot(z)
```

```python
ax = plt.subplot()
lines = ax.plot(x,y)   
ax.scatter(x,y)

plt.show()
```

```python
ax = plt.subplot()
lines = ax.plot(x,y)             #Draw points with lines or markers connecting them 
ax.scatter(x,y)  
ax.fill(x,y,color='red')
plt.show()
```

```python
ax = plt.subplot()
lines = ax.plot(x,y)             #Draw points with lines or markers connecting them 
ax.scatter(x,y)  
ax.fill(x,y,color='red')         #Draw filled polygons 
ax.fill_between(x,y,color='yellow')  #Fill between y-values and 0
 
```

```python
fig5, axes = plt.subplots(nrows=2,ncols=2)
 
axes[0,0].bar([1,2,3],[3,4,5])   #Plot vertical rectangles (constant width)      
axes[1,0].barh([0.5,1,2.5],[0,1,2])   #Plot horiontal rectangles (constant height) 
axes[1,1].axhline(0.45)          #Draw a horizontal line across axes   
axes[0,1].axvline(0.65)          #Draw a vertical line across axes 
```

```python
fig5, axes = plt.subplots(nrows=2,ncols=1)
axes[0].arrow(0,0,0.5,0.5)     # Add an arrow to the axes 
plt.show()
```

```python

fig6, axes = plt.subplots(nrows=2,ncols=2)
axes[0,0].hist(y)           # Plot a histogram 
axes[0,1].boxplot(y)        # Make a box and whisker plot
axes[1,0].violinplot(y)     # Make a violin plot
axes[1,1].violinplot(z)   
plt.show()
```
